```python
#  Import and read the charity_data.csv.
charity_file_path = "/Users/sequoiaboubionmckay/Desktop/DeepLearningProject/Starter_Code/charity_data.csv"
application_df = pd.read_csv(charity_file_path)
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN'
application_df = application_df.drop(columns=["EIN", "AFFILIATION", "USE_CASE", "SPECIAL_CONSIDERATIONS"])
application_df
```

```python
name_counts = application_df["NAME"].value_counts()
name_counts[name_counts>5]
```

```python
names_to_replace = list(name_counts[name_counts <= 1].index)

for app in names_to_replace:
    application_df["NAME"] = application_df["NAME"].replace(app, "Other")
application_df["NAME"].value_counts()
```

```python
# Determine the number of unique values in each column.
unique_values = application_df.nunique()
unique_values
```

```python
# Look at APPLICATION_TYPE value counts for binning
value_counts = application_df["APPLICATION_TYPE"].value_counts()
value_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff_value = 528

application_types_to_replace = list(value_counts[value_counts < cutoff_value].index)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
class_counts = application_df["CLASSIFICATION"].value_counts()
class_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
filtered_class_value_counts = class_counts[class_counts > 1]
filtered_class_value_counts
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff_value = 1880
classifications_to_replace = list(class_counts[class_counts < cutoff_value].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df.head()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_dummies_df = pd.get_dummies(application_df)
application_dummies_df
```

```python
# Split our preprocessed data into our features and target arrays
X = application_dummies_df.drop(["IS_SUCCESSFUL"], axis="columns").values
y = application_dummies_df["IS_SUCCESSFUL"].values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_feature = len(X_train[0])
hidden_nodes_layers1 = 80
hidden_nodes_layers2 = 30
hidden_nodes_layers3 = 10
hidden_nodes_layers4 = 5
hidden_nodes_layers5 = 2

neural_network_model = tf.keras.models.Sequential()

# First hidden layer
neural_network_model.add(
    tf.keras.layers.Dense(units=hidden_nodes_layers1, input_dim=number_input_feature, activation="tanh")
    )

# Second hidden layer
neural_network_model.add(
    tf.keras.layers.Dense(units=hidden_nodes_layers2, activation="tanh")
    )

# Third hidden layer
neural_network_model.add(
    tf.keras.layers.Dense(units=hidden_nodes_layers3, activation="tanh")
    )


# Fourth hidden layer
neural_network_model.add(
    tf.keras.layers.Dense(units=hidden_nodes_layers4, activation="tanh")
    )


# Output layer
neural_network_model.add(
    tf.keras.layers.Dense(units=1, activation="tanh")
    )

# Check the structure of the model
neural_network_model.summary()
```

```python
# Compile the model
neural_network_model.compile(loss="binary_crossentropy", optimizer='adam', metrics="accuracy")
```

```python
# Train the model
fit_model = neural_network_model.fit(X_train_scaled, y_train, epochs=50)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = neural_network_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
neural_network_model.save("AlphabetSoupCharityC.h5")
```
